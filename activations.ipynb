```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5.0, 5.0, 100)
```

```python
# Linear
y = x
plt.plot(x, y)
plt.xlim(x.min(), x.max())
```

```python
# Sigmoid
y = 1 / (1 + np.exp(-x))
plt.plot(x, y)
plt.xlim(x.min(), x.max())
```

```python
# Relu
y = np.maximum(0, x)
plt.plot(x, y)
plt.xlim(x.min(), x.max())
```
